```python
#Bivarite analysis
#scatter plot with target feature
plt.figure(figsize=(15,30))
for i, column in enumerate(df.columns):
  plt.subplot(7,3,i+1)
  plt.scatter(df[column],df['generated_power_kw'])
  plt.title(f'{column} vs Generated Power')
  plt.xlabel(column)
  plt.ylabel('Generated Power')
plt.tight_layout()
plt.show()
```
